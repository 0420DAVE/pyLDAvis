# Visualizing GraphLab's LDA TopicModel with pyLDAvis

This is an example of how to use [`pyLDAvis`](https://github.com/bmabey/pyLDAvis) [helper funtions](https://pyldavis.readthedocs.org/en/latest/modules/API.html#module-pyLDAvis.graphlab) to visualzie a GraphLab Create Topic Model model. For our example model we will be extending the example provided by [GraphLab's own documenation](https://dato.com/products/create/docs/generated/graphlab.topic_model.create.html).

```python
import graphlab as gl
import pyLDAvis
import pyLDAvis.graphlab
```

```python
# turn on automatic rendering of visualizations
pyLDAvis.enable_notebook()
```

```python
docs = gl.SArray('http://s3.amazonaws.com/GraphLab-Datasets/nytimes')
```

```python
topic_model = gl.topic_model.create(docs, num_topics=15)
```

```python
pyLDAvis.graphlab.prepare(topic_model, docs)
```
