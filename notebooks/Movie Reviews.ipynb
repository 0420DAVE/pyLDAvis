# Visualizing the Movie Reviews model from the R LDAvis project

```python
cd ../
```

```python
import json
import pyLDAvis as vis
```

```python
with open('tests/data/movie_reviews_input.json', 'r') as j:
        data_input = json.load(j)
```

```python
movies_pd = vis.prepare(topic_term_dists=data_input['phi'], doc_topic_dists=data_input['theta'], \
                       doc_lengths=data_input['doc.length'], vocab=data_input['vocab'], \
                       term_frequency=data_input['term.frequency'], R=30, lambda_step = 0.01)
```

```python
vis.display(movies_pd)
```
